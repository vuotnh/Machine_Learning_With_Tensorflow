To create the test URL valid set, I had to write a separate python program, `vgg_valid_url.py` and end up using multiprocessing.Pool and map reduce to parallelize the URL checks since checking 2.6M URLs is a pain on any system and due to timeouts, etc., and all the problems with crawling the web, doing it sequentially takes expotentially longer. I ended up with success on a 48 core machine with n=48 proc for multiprocessing and splitting the 2.6M URLs into 54k chunks to check.
   2. After getting a valid URL dataset, I wrote a program separate of this notebook called `vgg_create_img.py` to take the valid URLs from the pandas data frame CSV I made called `vgg_face_full_urls.csv` and download them in parallel. This took days. It was more than downloading the images from the URLs, it was dealing with all the not a valid image, and other errors that arose during the downloading process, as I alluded to before. I'm not talking automatically I'll get to that in a bit - but just simply taking a given bounding box, which VGG Face gives you, and then cropping it into a new image of size `244x244` which is the size of the input for the VGG Face Network I was going to construct. To collect all of the 1.25M URLs as face cropped images (for the successful URLs that were actually images), I had to use a supercomputer, parallelize things, and have `~90Gb` of space available for the uncompressed dataset. The yield from the process was a folder, `vgg-face` with 2622 celebrity names e.g., `Paul_Sorvino` at the top level directories, and inside of each directory, a file named `N_244x244.png` where `N` was the `df.reset()` index number result after reindexing after subsetting to just the valid URLs, and ultimately where `N` was between `1` and on average `~477` samples per class.
   3. Then the _real_ fun began. So, I have all this image data. How to construct a dataset from it? I ended up reading a ton of tutorials, in particular [this excellent one from the Tensorflow folks on building a dataset from images](https://www.tensorflow.org/tutorials/load_data/images) and I got exposed to the [Tensorflow Dataset API](https://www.tensorflow.org/guide/datasets). Auto `shuffle()` function. And I finally could take advantage of all of the cool [CIFAR-10 Tensorflow Image Manipulation I saw](https://github.com/tensorflow/models/blob/master/tutorials/image/cifar10/cifar10_input.py#L58) when doing the CIFAR-10 tutorial. So I was able to build a TF Dataset of Randomly Cropped, Noised, Rotated and Flipped Images and use TF.Dataset API and its ability to generate `one_shot_iterator()` pipelines to create batches of images and associated labels automatically from disks to train on a subset of the VGG Face data. I concluded that I only really could do about 4 celebrities overall, especially since I wanted to get the training started on my local laptop. Batch size was an important selection, I was able to do `BATCH_SIZE=256` on my local laptop but when I took this to a supercomputer with lots of GPUs the GPUs ran out of memory with that `BATCH_SIZE` and I ended up having to use `BATCH_SIZE=128` to get it running there, along with some other optimizations I will describe below. The tradeoff was fine as progress was reported more often and the GPU sped everything up. The 4 celebrities I picked for the training all fit the mean of `~477 images per sample`. After training on two different supercomputers for over a day, one of them progress to ~500 epochs out of 1000 and I achieved (at present) a `70% Micro-ROC average` on the four classes `Betty_White, Kate_Beckinsale, Paul_Sorvino, Zoe_Saldana`. My further goal is to extend this to support all `2622` celebrities, but that will require more time on the supecomputer.
   4. So I created a `40x244x244x3` test data set, and associated `40x1` label set, and then I cache and save them as `.npy` files so you don't have to generate them again after the first time.
   5. I added optimizations that monitored the validation loss, and accuracy and tried to determine how well the model could generalize beyond the 70% ROC I was mentioning. What I eventually found out was that a) I had some bugs in my original model that got the 70% ROC average (I had typos that didn't connect the right layers to the other right ones, effectively cutting out the full network); b) At `244x244` for `IMAGE_SIZE` and only `~477` samples per class, I didn't have enough data for such a dense network. Thanks to Nishant the author, and really to [Thamme Gowda](http://twitter.com/thammegowda) my former MS student in the [USC Information Retrieval and Data Science (IRDS)](http://irds.usc.edu/) group and current PhD student at [USC Information Sciences Institute](http://isi.edu/) for the help in realizing this. After trying and failing to get a good generalizable model, I ended up changing the model to be a `64x64` input `IMAGE_SIZE` model and then to chop off the last 6 convolutional layers and one of the fully connected layers and then to reduce the size of the fc to `128` that I did keep. That, combined with batch normalization, dropout=`0.5` and with `L2 regularization`, led to a really good model. I was wondering too - looking at FaceNet and other things - I wasn't learning the biases for the convolutional layers - [was this required?](https://stackoverflow.com/questions/51959507/does-bias-in-the-convolutional-layer-really-make-a-difference-to-the-test-accura). I did a [train/test split of 70/30 on the ~477 images per class](https://stackoverflow.com/questions/42471570/how-to-split-documents-into-training-set-and-test-set), and then used the test set for validation loss/accuracy to see how generalized the model was. At the end of the day it still didn't perform great on the actual random test data URLs (40) that I collected (only 30% when measured), but the model did amazing on the unseen test data part of the VGG Face data I downloaded (got a `~97% ROC accuracy`). 
   6. What I really learned previously in taking the next step in this assignment is the value of `train/test/eval` which I never really came to fully appreciate starting out the book. Train is easy. That's what you optimize and build the model weights/parameters and learn on. I always wondered really what the difference between Test and Evaluate is, but not any more. You use Evaluate or `Validation` data to measure how well your algorithm is learning by periodically (e.g., every 5 epochs) measuring the accuracy and loss against data the network hasn't seen. A generalized model will not just see training accuracy gains and loss decreases, but will also track similarly on validation (or unseen) data. The `test` data then remains to validate the generalizability of the model after training (and validation) is complete. Some people transpose test and validate, it doesn't really matter. But I learned a lot about that in this project. [This is a great read](https://www.tensorflow.org/beta/guide/keras/training_and_evaluation) in Tensorflow and Keras that helps explain it.

```python
def rect_to_bb(rect):
    # take a bounding predicted by dlib and convert it
    # to the format (x, y, w, h) as we would normally do
    # with OpenCV
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y
 
    # return a tuple of (x, y, w, h)
    return (x, y, w, h)
```

```python
def process_celeb(name, path):
    # format is 
    # Each line represents:
    # id url left top right bottom pose detection_score curation
    # id: Integer id for an image 
    # url: The weblink for the image. 
    # [left top right bottom] the bounding box for an image. 
    # pose: frontal/profile (pose>2 signifies a frontal face while
    # pose<=2 represents left and right profile detection).
    # detection score: Score of a DPM detector.
    # curation: Whether this image was a part of final curated dataset
    
    df = pd.read_csv(path, sep=' ', names=['ID', 'URL', 'LEFT', 'TOP', 'RIGHT', 'BOTTOM', 'POSE', 'DETECTION', 'CURATION'])
    celeb_list = [name] * len(df)
    df['CELEB'] = celeb_list
    return df
        
```

```python
def preprocess_image(image, distort=True):
    image = tf.image.decode_png(image, channels=3)
    image = image[..., ::-1] # images are stored on files with these reversed, so fix per https://stackoverflow.com/questions/42161916/tensorflow-how-to-switch-channels-of-a-tensor-from-rgb-to-bgr  
    image = tf.image.resize(image, [IMAGE_SIZE, IMAGE_SIZE])

    if distort:
        # Randomly flip the image horizontally.
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_brightness(image, max_delta=63)
        image = tf.image.random_contrast(image, lower=0.2, upper=1.8)

        rotate_pct = 0.5 # 50% of the time do a rotation between 0 to 90 degrees
        if random.random() < rotate_pct:
            degrees = random.randint(0, 90)
            image = tfa.image.rotate(image, degrees * math.pi / 180, interpolation='BILINEAR')
    
        # Fixed standardization 
        image = (tf.cast(image, tf.float32) - 127.5)/128.0
        # Subtract off the mean and divide by the variance of the pixels.
    
    image = tf.image.per_image_standardization(image)

    return image
```

```python
def get_training_and_testing_sets(file_list):
    split = 0.7
    split_index = math.floor(len(file_list) * split)
    training = file_list[:split_index]
    testing = file_list[split_index:]
    return training, testing
```

```python
def predict(img_data, ckpt, save_path, noise=False):
    class_num, class_name, confidence = None, None, 0.

    image = img_data

    if noise:
        # Randomly flip the image horizontally.
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_brightness(image, max_delta=63)
        image = tf.image.random_contrast(image, lower=0.2, upper=1.8)

        # Subtract off the mean and divide by the variance of the pixels.
        image = tf.image.per_image_standardization(image)

    im_data = tf.reshape(image, [1, IMAGE_SIZE, IMAGE_SIZE, 3])
    im_data = im_data.numpy()

    ckpt.restore(save_path).assert_consumed()
    print("Model restored.")
    logits = model(im_data, training=False)
    class_num = np.argmax(logits, axis=1)[0]
    class_name = label_names[class_num]
    confidence = logits[0,class_num].numpy()
    all_preds = logits.numpy()
            
            
    return (class_num, class_name, confidence, all_preds)
```

```python
def get_test_accuracy(test_data, test_names, test_labels, ckpt, save_path):
    class_num, class_name, confidence = None, None, 0.
    ckpt.restore(save_path).assert_consumed()
    print("Model restored.")
    
    onehot_test_labels = tf.one_hot(test_labels, len(test_names), on_value=1., off_value=0., axis=-1)  
    logits = model(test_data,training=False)
    test_logits_out = tf.nn.softmax(logits)      
    test_correct_pred = tf.equal(tf.argmax(test_logits_out, 1), tf.argmax(onehot_test_labels, 1))
    test_accuracy = tf.reduce_mean(tf.cast(test_correct_pred, tf.float32))

    print('Test accuracy %f' % (test_accuracy.numpy()))  
    predictions = tf.argmax(test_logits_out, 1).numpy()
    return (predictions, tf.cast(test_correct_pred, tf.float32).numpy(), onehot_test_labels, test_logits_out)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
```

```python
def loss(inputs, outputs):
    return tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=inputs, labels=outputs))
```

```python
def train_step(model, inputs, outputs):
    with tf.GradientTape() as t:
        current_loss = loss(model(inputs), outputs)

    grads = t.gradient(current_loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads,model.trainable_variables))
    return current_loss
```

```python
def correct_predictions(model, inputs, outputs):
    return tf.equal(tf.argmax(model(inputs), 1), tf.argmax(outputs, 1))
```

```python
def accuracy(correct_pred):
    return tf.reduce_mean(tf.cast(correct_pred, tf.float32), name='accuracy')
```

```python
class VGGFaceModel(tf.keras.Model):

    def __init__(self, training=False, num_cls=4):
        super(VGGFaceModel, self).__init__()
        self.num_cls = num_cls
        self.training = training
        
        self.zp2d_1 = tf.keras.layers.ZeroPadding2D((1,1),input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3))
        self.cv2d_1_1 = tf.keras.layers.Conv2D(64, (3,3), activation='relu')
        self.cv2d_1_2 = tf.keras.layers.Conv2D(64, (3,3), activation='relu')
        self.mp2d_1 = tf.keras.layers.MaxPooling2D((2,2), strides=(2,2))
        self.bn_1 = tf.keras.layers.BatchNormalization()
        
        self.zp2d_2 = tf.keras.layers.ZeroPadding2D((1,1))
        self.cv2d_2_1 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu')
        self.zp2d_3 = tf.keras.layers.ZeroPadding2D((1,1))
        self.cv2d_2_2 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu')
        self.mp2d_2 = tf.keras.layers.MaxPooling2D((2,2), strides=(2,2))
        self.bn_2 = tf.keras.layers.BatchNormalization()

        
        self.zp2d_4 = tf.keras.layers.ZeroPadding2D((1,1))
        self.cv2d_3_1 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu')
        self.mp2d_3 = tf.keras.layers.MaxPooling2D((2,2), strides=(2,2))
        self.bn_3 = tf.keras.layers.BatchNormalization()

        self.flatten = tf.keras.layers.Flatten()
        self.dense_1 = tf.keras.layers.Dense(128, activation="relu")
        self.dropout_1 = tf.keras.layers.Dropout(0.5)
        self.bn_4 = tf.keras.layers.BatchNormalization()

        
        self.out = tf.keras.layers.Dense(self.num_cls, activation="softmax")

        
    def call(self, x):
        
        x = self.zp2d_1(x)
        x = self.cv2d_1_1(x)
        x = self.cv2d_1_2(x)
        x = self.mp2d_1(x)
        x = self.bn_1(x, training=self.training)
        
        x = self.zp2d_2(x)
        x = self.cv2d_2_1(x)
        x = self.zp2d_3(x)
        x = self.cv2d_2_2(x)
        x = self.mp2d_2(x)
        x = self.bn_2(x, training=self.training)

        x = self.zp2d_4(x)
        x = self.cv2d_3_1(x)
        x = self.mp2d_3(x)
        x = self.bn_3(x, training=self.training)
        
        x = self.flatten(x)
        x = self.dense_1(x)
        x = self.dropout_1(x)        
        x = self.bn_4(x, training=self.training)
        x = self.out(x)    
        
        return x
```

```python
print('batch size', BATCH_SIZE)
for j in tqdm(range(0, epochs)):
    iter = list(train_ds.as_numpy_iterator())
    val_iter = list(val_ds.as_numpy_iterator())

    assert len(val_iter) == 1
    val_image_batch, val_label_batch = val_iter[0]
    onehot_val_labels = tf.one_hot(val_label_batch, len(label_names), on_value=1., off_value=0., axis=-1)

    for (image_batch, label_batch) in iter:
        onehot_labels = tf.one_hot(label_batch, len(label_names), on_value=1., off_value=0., axis=-1)
        err = train_step(model, image_batch, onehot_labels)
        
    ckpt.step.assign_add(1)

    correct_pred = correct_predictions(model, image_batch, onehot_labels)
    acc = accuracy(correct_pred)

    print("Step %d Train Accuracy %f Train Loss %f " % (j, acc.numpy(), err.numpy()))

    if j != 0 and j % 5 == 0: #every 5 steps validate and use to set rate
        vcp = correct_predictions(model, val_image_batch, onehot_val_labels)
        v_acc = accuracy(vcp)
        v_err = loss(model(val_image_batch), onehot_val_labels)

        print("Step %d Validation Accuracy %f Validation Loss %f" % (j, v_acc.numpy(), v_err.numpy()))     
        
    if int(ckpt.step) % 50 == 0:
        save_path = manager.save()
        print("Saved checkpoint for step {}: {}".format(int(ckpt.step), save_path))
        print("loss {:1.2f}".format(err.numpy()))

save_path = manager.save()
print("Saved checkpoint for step {}: {}".format(int(ckpt.step), save_path))
print("loss {:1.2f}".format(err.numpy()))
```
